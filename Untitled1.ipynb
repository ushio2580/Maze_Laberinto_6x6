#  Import required libraries


```python
# Cell 1: Import and setup
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from IPython.display import display, Markdown, Math, Latex

# Configure display settings
sp.init_printing(use_latex='mathjax')
%matplotlib inline
plt.style.use('ggplot')
plt.rcParams['figure.figsize'] = (12, 8)
plt.rcParams['axes.labelsize'] = 12
```

```python
# Cell 2: Define display helper function
def show(title, content=None, math=None):
    """Display formatted content"""
    display(Markdown(f"## {title}"))
    if content:
        display(Markdown(content))
    if math:
        display(Math(math))
```

```python
# Cell 3: Example 1 - Gradient and Hessian
show("üßÆ Example 1: Gradient and Hessian", 
     "### Goal: Analyze a function of two variables x‚ÇÅ and x‚ÇÇ")

# Define symbols and default function
x1, x2 = sp.symbols('x1 x2')
user_input = "1/2*(x2 - x1**2)**2 + (1 - x1)**2"
f_expr = sp.sympify(user_input)

show("Function Definition:", 
     math=r"f(x_1, x_2) = \frac{1}{2}(x_2 - x_1^2)^2 + (1 - x_1)^2")

# Calculate first derivatives
grad_x1 = sp.diff(f_expr, x1)
grad_x2 = sp.diff(f_expr, x2)

show("First Partial Derivatives:",
     math=(r"\begin{aligned}"
          r"\frac{\partial f}{\partial x_1} &= " + sp.latex(grad_x1) + r"\\"
          r"\frac{\partial f}{\partial x_2} &= " + sp.latex(grad_x2) +
          r"\end{aligned}"))

# Calculate gradient
gradient = sp.Matrix([grad_x1, grad_x2])
show("Gradient Vector:", 
     math=r"\nabla f = " + sp.latex(gradient))

# Calculate Hessian
hessian = sp.hessian(f_expr, (x1, x2))

# Display second partial derivatives in a single formatted block
show("Second Partial Derivatives:",
     math=(r"\begin{aligned}"
           r"\frac{\partial^2 f}{\partial x_1^2} &= " + sp.latex(hessian[0,0]) + r" \\"
           r"\frac{\partial^2 f}{\partial x_1 \partial x_2} &= " + sp.latex(hessian[0,1]) + r" \\"
           r"\frac{\partial^2 f}{\partial x_2 \partial x_1} &= " + sp.latex(hessian[1,0]) + r" \\"
           r"\frac{\partial^2 f}{\partial x_2^2} &= " + sp.latex(hessian[1,1]) +
           r"\end{aligned}"))

show("Hessian Matrix:", 
     math=r"\nabla^2 f = " + sp.latex(hessian))
```

```python
# Cell 4: Example 1 - 3D Visualization
f_num = sp.lambdify((x1, x2), f_expr, 'numpy')

# Create grid with proper bounds
x = np.linspace(-2, 2, 50)
y = np.linspace(-1, 3, 50)
X, Y = np.meshgrid(x, y)

# Handle potential numerical issues
with np.errstate(all='ignore'):
    Z = f_num(X, Y)
    Z = np.nan_to_num(Z, nan=0, posinf=0, neginf=0)

fig = plt.figure(figsize=(14, 6))

# 3D Surface Plot
ax1 = fig.add_subplot(121, projection='3d')
surf = ax1.plot_surface(X, Y, Z, cmap='viridis', 
                       rstride=1, cstride=1,
                       linewidth=0, antialiased=True,
                       alpha=0.8)
ax1.set_title("3D Surface Plot")
ax1.set_xlabel(r"$x_1$")
ax1.set_ylabel(r"$x_2$")
ax1.set_zlabel(r"$f(x_1,x_2)$")

# Contour Plot
ax2 = fig.add_subplot(122)
contour = ax2.contour(X, Y, Z, levels=20, cmap='viridis')
ax2.set_title("Contour Plot")
ax2.set_xlabel(r"$x_1$")
ax2.set_ylabel(r"$x_2$")
plt.colorbar(contour, label='Function Value')

plt.tight_layout()
plt.show()
```

```python
# Cell 5: Example 2 - Line Search
show("üìà Example 2: Line Search Optimization",
     "### Goal: Find minimum along a specified direction")

# Define function and parameters
f_line = x1**2 + x2**2 + x1*x2 + 4*x1 - x2 + 1
x0 = sp.Matrix([2, -1])
p = sp.Matrix([-1, 1])

show("Problem Setup:",
     math=(r"\begin{aligned}"
          r"f(x_1,x_2) &= " + sp.latex(f_line) + r"\\"
          r"\vec{x}_0 &= " + sp.latex(x0) + r"\\"
          r"\vec{p} &= " + sp.latex(p) +
          r"\end{aligned}"))

# Define parameterized path
t = sp.symbols('t')
x_t = x0 + t * p
f_t = f_line.subs({x1: x_t[0], x2: x_t[1]})

show("Path Definition:",
     math=r"\vec{x}(t) = \vec{x}_0 + t\vec{p} = " + sp.latex(x_t))
show("Function Along Path:",
     math=r"f(t) = " + sp.latex(f_t))
```

```python
# Cell 6: Find Optimal Point
f_t_deriv = sp.diff(f_t, t)
t_opt = sp.solve(f_t_deriv, t)[0]
z = x_t.subs(t, t_opt)

show("Optimal Point Calculation:",
     math=(r"\begin{aligned}"
          r"f'(t) &= " + sp.latex(f_t_deriv) + r"\\"
          r"t^* &= " + sp.latex(t_opt) + r"\\"
          r"\vec{z} &= " + sp.latex(z) +
          r"\end{aligned}"))

# Verify optimality condition
grad_f = sp.Matrix([sp.diff(f_line, x1), sp.diff(f_line, x2)])
grad_z = grad_f.subs({x1: z[0], x2: z[1]})
dot_product = (grad_z.T * p)[0]

show("Optimality Verification:",
     math=(r"\begin{aligned}"
          r"\nabla f(\vec{z}) &= " + sp.latex(grad_z) + r"\\"
          r"\nabla f(\vec{z})^T \vec{p} &= " + sp.latex(dot_product) +
          r"\end{aligned}"))

if dot_product == 0:
    display(Markdown("**‚úÖ Verification Passed:** The directional derivative is zero (minimum found)"))
else:
    display(Markdown("**‚ö†Ô∏è Warning:** Dot product not zero (may not be minimum)"))
```
